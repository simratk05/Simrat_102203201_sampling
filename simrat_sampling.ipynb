```python
import pandas as pd

sampling_methods = ['SMOTE', 'ROS', 'RUS', 'TMK', 'NM']
models = ['Decision tree', 'Random forest', 'Logistic Regression', 'Linear regression', 'SVM']

# Create an empty DataFrame with the specified rows and columns
df = pd.DataFrame(index=models, columns=sampling_methods)
```

# 1. SMOTE

```python
from imblearn.over_sampling import SMOTE
```

```python
x=data.drop('Class', axis=1)
y=data['Class']
```

```python
x_res,y_res =SMOTE().fit_resample(x,y)
```

```python
y_res.value_counts()
```

**here sampling is done with class 1 and class 0 samples having 756 records each**

```python
x_train,x_test, y_train, y_test=train_test_split(x_res,y_res,test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier 
from sklearn.tree import DecisionTreeClassifier 
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, f1_score ,mean_squared_error
```

```python
#decision tree
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
y_pred= model.predict(x_test)
accuracy= accuracy_score(y_test,y_pred)
print("Decision Tree:")
print(f"Accuracy:{accuracy}")
f1= f1_score(y_test,y_pred)
print(f"f1_score:{f1}")
df.loc['Decision tree', 'SMOTE'] = accuracy
```

```python
#random foreest
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(x_train, y_train)
rf_pred = rf_model.predict(x_test)
rf_accuracy = accuracy_score(y_test, rf_pred)
rf_f1 = f1_score(y_test, rf_pred, pos_label=1)

print("Random Forest:")
print(f"Accuracy: {rf_accuracy}")
print(f"F1 Score: {rf_f1}\n")
df.loc['Random forest', 'SMOTE'] = rf_accuracy
```

```python
# Logistic Regression
log_reg_model = LogisticRegression(max_iter=100, random_state=42)
log_reg_model.fit(x_train, y_train)
log_reg_pred = log_reg_model.predict(x_test)
log_reg_accuracy = accuracy_score(y_test, log_reg_pred)
log_reg_f1 = f1_score(y_test, log_reg_pred, pos_label=1)

print("Logistic Regression:")
print(f"Accuracy: {log_reg_accuracy}")
print(f"F1 Score: {log_reg_f1}\n")
df.loc['Logistic Regression', 'SMOTE'] = log_reg_accuracy
```

```python
# Linear Regression (for regression tasks)
lin_reg_model = LinearRegression()
lin_reg_model.fit(x_train, y_train)
lin_reg_pred = lin_reg_model.predict(x_test)
lin_reg_mse = mean_squared_error(y_test, lin_reg_pred)

print("Linear Regression:")
print(f"Mean Squared Error: {lin_reg_mse}\n")
```

```python
# Support Vector Machine (SVM)
svm_model = SVC(kernel='linear', random_state=42)
svm_model.fit(x_train, y_train)
svm_pred = svm_model.predict(x_test)
svm_accuracy = accuracy_score(y_test, svm_pred)
svm_f1 = f1_score(y_test, svm_pred, pos_label=1)

print("Support Vector Machine:")
print(f"Accuracy: {svm_accuracy}")
print(f"F1 Score: {svm_f1}")
df.loc['SVM', 'SMOTE'] = svm_accuracy
```
